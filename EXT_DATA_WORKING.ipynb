# Преобразование для векторной маски

Теперь наша задача сформировать запрос для наиболее релевантных игр и изданий

```python
external_dataset.head()
```

```python
publisher_list_df_sales = external_dataset.pivot_table(index = ['Publisher'], values = 'Total_Shipped', aggfunc=np.sum).sort_values('Total_Shipped',ascending=False).reset_index()

publisher_list_df_sales.head(30)
```

```python
publisher_list_df_quantity = external_dataset.pivot_table(index = ['Publisher'], values = 'Name', aggfunc='count').sort_values('Name',ascending=False).reset_index()
publisher_list_df_quantity.head(30)
```

```python
publishers_1 = set(publisher_list_df_sales['Publisher'].values[:40])

publishers_2 = set(publisher_list_df_quantity['Publisher'].values[:40])

publishers_set = publishers_1.union(publishers_2)

publishers_set.remove('Unknown')

print(len(publishers_set))

print(sorted(publishers_set))
```

```python
games_reduced = external_dataset.query('Publisher in @publishers_set')
games_names = list(games_reduced.Name.values)
```

```python
d_to_parse = dict()

d_to_parse['publishers'] = sorted([publisher for publisher in publishers_set])
d_to_parse['games'] = sorted([name for name in games_names])
d_to_parse
```

```python
with open('queries_iter_channels.json', 'w', encoding='utf-8') as f:
    json.dump(d_to_parse, f)
```
